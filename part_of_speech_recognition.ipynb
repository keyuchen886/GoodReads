```python
import nltk
import json
import matplotlib.pyplot as plt

from wordcloud import WordCloud, ImageColorGenerator

import pickle

from matplotlib.backends.backend_pdf import PdfPages
```

```python
REVIEW_DIR = ".\data\goodreads_reviews_comics_graphic.json"
```

```python
def extract_adj(sent):
    token = nltk.word_tokenize(sent)
    pos_tag = nltk.pos_tag(token)
    adj = [ x[0] for x in pos_tag if x[1]=="JJ"]
    return(adj)
```

```python
num_records = 529532  #there are 529532 record in total
index = 0 
#num_records=40
books = {}

#----------------------------
#     run main
#----------------------------

with open(REVIEW_DIR) as fie:
    for review in fie:
        
        if index > num_records:
            fie.close()
            break
            
        
            
        record = json.loads(review)  #load json as a dictionary
        adj = extract_adj( record['review_text'] )
        try:
            books[record['book_id']]+=adj
        except:
            books[record['book_id']]=adj

            
        
        #print(i)
        index+=1
    
    
    
    fie.close()
```

```python
with open("./data/book_embedding.firefire",'rb') as file:
    tab = pickle.load(file)
    file.close()
```

```python
pp = PdfPages('multipage.pdf')

for k,v in books.items():
    if len(v)>=1:
        try:
            title = tab[k]['name']
            text=" ".join(v)
            wordcloud= WordCloud( background_color="white", max_words=1000).generate(text)
            fig = plt.figure()
            plt.imshow(wordcloud, interpolation="bilinear")
            plt.axis("off")
            plt.title( title)
            #plt.show()
            plt.savefig(pp, format='pdf')
        except:
            continue
pp.close()
```
